```python
#把同義詞建成字典
test_file1 = open('ingredient_lists.txt','r')
dic={}
lines_counter = 0
for i in test_file1.readlines():
    if test_file1 != '\n':
        lines_counter += 1
    if '' in i.split(','): 
        a = i.split(',')[:i.split(',').index('')] #如果有空白的話，把後面空白截掉
        for j in a[1:]:
            dic[j] = a[0] #把同義詞對到食材
    else:
        a = i.split(',') 
        for j in a[1:]:
            dic[j] = a[0]
            
print '食材tag數量： ', lines_counter
QQQ = len(dic)-lines_counter
print '食材（含同義字）數量： ', QQQ
```

```python
#skip duplicate 篩檔案完全重複的row
in_file = open('All.txt','r')
out_file = open('All-no-dup.txt','w')
unique = set()
for line in in_file.readlines():
    if line in unique: 
        continue 
    else:
        unique.add(line)
        out_file.write(line)
in_file.close()
out_file.close()

#把第每一欄的東東切乾淨
in_file = open('All-no-dup.txt','r')
out_file = open('All-clean.txt','w')
for i in in_file.readlines():
    a = i.split(',')[0]
    c = i.split(',',1)[1]
    if '（' in a:
        b = a.split('（')[0]
        out_file.write((b+','+c))
    elif '(' in a:
        b = a.split('(')[0]
        out_file.write((b+','+c))
    else:
        out_file.write(i)
        
in_file.close()
out_file.close()
```

```python
#calculate kal per unit(g)
import re
in_file = open('All-clean.txt','r')
out_file = open('Calories_table1.txt','w')

for i in in_file.readlines():
    q1 = i.split(',')[0].strip()
    q2 = i.split(',')[1].strip()
    q3 = i.split(',')[2].strip()
    q4 = i.split(',')[3].strip()
    Amm = re.match(r'(?P<num>\d+).*',q3)
    Bmm = re.match(r'(?P<num>\d+).*',q4)
    if (Amm is not None) & (Bmm is not None):
        qq3 = Amm.group("num") 
        qq4 = Bmm.group("num")
        q5 = float(qq4)/float(qq3)
        out_file.write(str(q1) +','+ str(q2) +','+ str(qq3) +','+ str(qq4) +','+ str(q5) +'\n')
    else:
        out_file.write(str(q1) +','+ str(q2) +','+ str(q3) +','+ str(q4) +','+'\n')
in_file.close()
out_file.close()
```

```python
in_file = open('Calories_table1.txt','r')
count = {}

for i in in_file.readlines():
    a = i.split(',',4)[0]
    if a in count:
        count[a] = count[a] + 1
    else:
        count[a] = 1
in_file.close()

dup=[]        
for k in count:
    #print k, count[k]
    if count[k] >1:
        dup.append(k)

out_file = open('ingre_dups.txt','w')
for g in dup:
    in_file = open('Calories_table1.txt','r')
    for j in in_file.readlines():
        if g == j.split(',')[0]:
            out_file.write(j)
        in_file.close()
out_file.close()
            

```

```python
#myfitnesspal
import requests
from bs4 import BeautifulSoup
myfitnesspal = open('myfitnesspal.txt', 'w')
page_format= ('http://www.myfitnesspal.com.tw/food/update_nutrition_facts/%d')
for page in range(100000000, 200000000):
    res = requests.get(page_format%(page))
    soup = BeautifulSoup(res.text) 
    try:
        if soup.select('.food-description')[0].text is not None:
            Qaa = soup.select('.food-description')[0].text
            Qbb = soup.select('select#food_entry_weight_id option')[0].text
            Qcc = soup.select('#nutrition-facts .col-2')[2].text
            myfitnesspal.write(Qaa+','+Qbb+','+Qcc+'\n')
    except:
        pass
myfitnesspal.close()
```
